# Señales Aleatorias

## Muestreo aleatorio

Señales aleatorias se puede simular a partir de generadores aleatorios, numpy tiene un generador aleatorio, la documentación se encuentra en este link [numpy.random](https://numpy.org/doc/stable/reference/random/index.html?highlight=random#module-numpy.random). A partir de sus metodos, como standard_normal() se pueden obtener muestras de diferente distribuciones de probabilidad

La [distribución de probabilidad normal estandar](https://es.wikipedia.org/wiki/Distribuci%C3%B3n_normal) está definida por la siguiente ecuación


$$\varphi(x) = \frac{e^{-\frac{x^2}{2}}}{\sqrt{2\pi}}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import default_rng

sampling_rate = 1000 # Hz
# n es la variable de tiempo discreto
n  = np.arange(0,3,1/sampling_rate)
N     = len(n)

# noise level, measured in standard deviations
noiseamp = 5

random_values = np.random.standard_normal(N)

# A more advanced way of generating random numbers 
# with better statistical properties 
rng = default_rng()
other_random_values = rng.standard_normal(N)


noise_signal = noiseamp * other_random_values

plt.plot(n,noise_signal,label='noise')


plt.legend()
plt.xlabel('Time (sec.)')
plt.ylabel('Amplitude')
plt.title('Noise signal')

plt.show()
```

```python
plt.hist(noise_signal, bins='auto')  # arguments are passed to np.histogram
```
