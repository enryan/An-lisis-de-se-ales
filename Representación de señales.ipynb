### Viariable discreta

$$x[n] = A \cos (\Omega_0 n + \theta)$$

Esta es una señal **determinística** completamente caracterizada por tres parámetros: $A$ es la amplitud de la cosenosoide, $\Omega$ es la frecuencia en radianes por segundo (rad/s), y $\theta$ es la fase en radianes. En lugar de $\Omega$, a menudo se utiliza la frecuencia $F$ ciclos por segundo o Hercio (Hz), donde $\Omega = 2 \pi F$  $\Omega = 2 \pi * 1/T$</math>..

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import pi, sin, cos
from math import radians

# n es una variable en el tiempo de -16.666 ms hasta 16.666ms
n = np.arange(-0.01666/2, 0.01666/2, 0.0001)
print("Número de muestras (N) = " + str(len(n)))
#print(n)
#Para una frecuencia de 60 Hz
Omega = 2* pi * 60 # 2*pi rad * 60 Hz(1/s) = rad/s
A = 1
theta = 0
# cos(n) no es una variable, cos(n) es aplicar la función cos a los valores en la variable n
x = A*cos(Omega*n + theta) # Unidades
#print(coseno)
#y2 = cos(n)
i_seno = A* cos(Omega*n - pi/2)
i_offset = i_seno + 2
señal_atrasada = cos(Omega*n - radians(153))

plt.plot(n, x, 'r')
plt.plot(n, i_seno, 'g')
plt.plot(n, señal_atrasada, 'b')
plt.grid()

v_seno = A* cos(Omega*(0.0083) - pi/2)
print(v_seno)
```

# Valor cuadratico medio o RMS (root mean square)

El valor cuadratico medio para una señal en tiempo discreto de N valores {$x_1$, $x_2$, ... , $x_n$,} de una variable discreta $x$, viene dada por la fórmula:


$$X_{\mathrm{RMS}} = 
\sqrt {{1 \over N} \sum_{i=1}^{N} x_i^2} =
\sqrt {{x_1^2 + x_2^2 + \cdots + x_n^2} \over N}
\qquad\qquad
$$

**Importante:** El rms de una señal periodica debe calcularse en un periodo de la señal (periodo fundamental)

```python
#print(x)
rms = np.sqrt(np.mean( i_offset**2 ))
print(rms)
```

# Representación fasorial

La transformada fasorial o representación fasorial permite cambiar de forma trigonométrica a forma compleja:

$$V_m e^{j \phi } = \mathcal{P} \{ V_m \cos( \omega t + \phi ) \}$$

donde la notación $\mathcal{P} \{x\}$ se lee como "transformada fasorial de x" 

Por ejemplo, el fasor del una señal sinosoidal periodica $V_m \cos( \omega t - \pi/2 )$ es $V_m (-\pi/2)$


## Corriente en atraso

Una señal de corriente se dice que está en atraso con respecto a otra señal de voltaje cuando la diferencia entre el ángulo del fasor de voltaje y el ángulo del fasor de corriente es positiva y está en adelanto cuando la diferencia entre los ángulos es negativa

La corriente en el circuito RL con respecto al voltaje está efectivamente en atraso.

Desde fasores nuestra fuente de voltaje sinusoidal es V (-90). 

Si la carga es RL, el ángulo de la representación de Z en coordenadas polares es positivo, por ejemplo para un Z con ángulo 63 la corriente seria I (-153)

La diferencia entre las fases de los fasores de voltaje y corriente es dif = -90 - ( - 153) = -90 + 153, que es equivalente a 63 positivo, por lo tanto la corriente esta en atraso con respecto al voltaje.
